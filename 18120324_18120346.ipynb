```python
#import thư viện
import numpy as np
import PIL
import matplotlib.pyplot as plt
from scipy.spatial.distance import cdist

#Các kích thước của ảnh gốc
img=PIL.Image.open('Lenna_(test_image).png')
img_1d=np.array(img)

img_height=img_1d.shape[0]
img_width=img_1d.shape[1]
num_channels=img_1d.shape[2]

img_1d=np.reshape(img_1d,(img_width*img_height,num_channels))

#các hàm phụ kmeans

#hàm tạo centroids ban đầu
def kmeans_init_centroids(img_1d, k_clusters, num_channels, init_centroids):
    if init_centroids=='random':
        return np.random.randint(0, 256, size=(k_clusters,num_channels))
    else:
        if init_centroids=='in_pixels':
            return img_1d[np.random.choice(img_1d.shape[0], k_clusters, replace=False)]

#hàm cập nhật nhãn
def kmeans_assign_labels(img_1d, centroids):
    labels=[]
    distance=cdist(img_1d,centroids)
    for pos in range(img_1d.shape[0]):
        labels.append(np.argmin(distance[pos]))
    return labels

#hàm cập nhật centroids
def kmeans_update_centroids(img_1d, labels, k_clusters, num_channels):
    new_centroids=np.zeros((k_clusters,num_channels), dtype = int)

    for k in range(k_clusters):
        color_k=[]
        for color, label in zip(img_1d,labels):
            if label==k:
                color_k.append(color)
        new_centroids[k]=np.mean(color_k, axis = 0).astype(int)
    return new_centroids

#hàm hiển thị ảnh mới sau khi giảm màu
def kmeans_show_new_image(centroids, labels):
    new_img=[]
    for label in labels:
        new_img.append(centroids[label])
    new_img=np.reshape(new_img,(img_height,img_width,num_channels))
    plt.imshow(new_img)
    plt.show()

#hàm chính kmeans
def kmeans(img_1d, k_clusters, num_channels,init_centroids='random'):
    centroids=kmeans_init_centroids(img_1d, k_clusters, num_channels, init_centroids)
    labels=[]
    
    while True:
        labels=kmeans_assign_labels(img_1d,centroids)
        new_centroids=kmeans_update_centroids(img_1d,labels,k_clusters,num_channels)
        if(np.array_equal(new_centroids,centroids)):
            break
        centroids=new_centroids
        
    return (centroids,labels)

#vòng lặp hiển thị ảnh với k=3,5,7
for k_clusters in [3,5,7]:
    (centroids,labels)=kmeans(img_1d,k_clusters,num_channels,'in_pixels')
    kmeans_show_new_image(centroids, labels)
```
